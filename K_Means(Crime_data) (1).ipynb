```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("crime_data.csv")
```

```python
df
```

```python
df1 = df.rename(columns = {'Unnamed: 0' : 'States'})
```

```python
df1
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(df1.iloc[:,1:])
```

```python
scaled_crime_df
```

```python
wcss = []#Elbow curve to choose the optimum k-value of cluster for crime data
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(5, random_state=42)#Clustering based on k=5 coz curve doesnt change much after that
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
df1['clusterid_new'] = clusters_new.labels_
```

```python
df1
```

```python
df1.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
# from this we can infer the overall crimes for states in cluster 1 are relatively low
```
